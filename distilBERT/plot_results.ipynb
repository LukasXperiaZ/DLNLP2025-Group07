```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Define your data ---
# Replace these with your actual model names and their corresponding SST-2 accuracies.
# Accuracies should be between 0.0 and 1.0.
model_names = [
    "Pretrained HF",
    "Wiki 6L, 1e",
    "Wiki 4L, 1e",
    "Wiki 4L, 2e"
]

accuracies = [
    0.903,  
    0.533,
    0.500, 
    0.507,
]

# Ensure the number of models matches the number of accuracies
if len(model_names) != len(accuracies):
    raise ValueError("The number of model names must match the number of accuracies.")

# --- 2. Create the bar plot ---
plt.figure(figsize=(6, 6)) # Adjust figure size as needed for readability
bars = plt.bar(model_names, accuracies, color=['skyblue', 'gray', 'lightgreen', 'green'])

# --- 3. Add labels and title ---
plt.xlabel("Model", fontsize=12)
plt.ylabel("Accuracy", fontsize=12)
plt.title("SST-2 Classification Accuracy (base DistilBERT models listed)", fontsize=14)

# --- 4. Set Y-axis limits (important for clarity) ---
plt.ylim(0.4, 1.0) # To zoom in on high accuracies if differences are small

# --- 5. Add accuracy values on top of the bars (optional, but good for clarity) ---
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.005, round(yval, 3), ha='center', va='bottom', fontsize=10)

# --- 6. Rotate x-axis labels if they overlap ---
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability if they are long
plt.tight_layout() # Adjust layout to prevent labels from being cut off

plt.savefig("sst2_accuracies.png", dpi=300)

# --- 7. Display the plot ---
plt.show()
```
